```python
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

```python
%%time
window =  2
step = 0
# moves = 100
for window in range(2, 50):
    profit = 0
    all_money = 100
    kol_good = 0
    buy, sell = [], []
    action = 1 # 1|-1
    moves = 0.9*all_money
    last_moves = 0
    minn, maxx = None, None
    for i in range(step+2*window-2, step+len(arr[:44000])):
        assert all_money > 0, (color.RED + 'Error', all_money, window, i)
        # mean = np.mean(arr[i-window+1:i+1])
        if minn is None:
            minn = np.min(arr[i-window+1:i+1])
            maxx = np.max(arr[i-window+1:i+1])
            continue
        lmin, lmax = minn, maxx
        minn = np.min(arr[i-window+1:i+1])
        maxx = np.max(arr[i-window+1:i+1])
        if lmax < maxx and action == -1:
            for k in sell[:1]:
                sell.remove(k)
                profit += k - arr[i] - 0.0001
                all_money += moves*(1 + (k - arr[i]))
                kol_good += 1
            action = 1
            last_moves = -moves
            moves = 0.9*(all_money)/arr[i]
            buy.append(arr[i])
            all_money -= moves*arr[i]
        if lmin > minn and action == 1:
            for k in buy[:1]:
                buy.remove(k)
                profit += arr[i] - k - 0.0001
                all_money += moves*arr[i]
                kol_good += 1
            action = -1
            last_moves = moves
            moves = all_money*0.9
            sell.append(arr[i])
            all_money -= moves
    for k in buy:
        if last_moves < 0:
            all_money += moves*arr[-1]
        else:
            all_money += last_moves*arr[-1]
    for k in sell:
        if last_moves < 0:
            all_money += last_moves*(1 + (k - arr[-1]))
        else:
            all_money += moves*(1 + (k - arr[-1]))
    print(color.END + f'window = {window}\n\t' + color.UNDERLINE + 'profit =' + color.BOLD + f' {round(profit, 4) + sum(arr[-1] - buy) + sum(sell - arr[-1])}')
    print(color.END + color.UNDERLINE + '\tBuy = ', sum(arr[-1] - buy) , 'Sell = ',  sum(sell - arr[-1]))
    print('\tKol_buy = ', len(buy), 'Kol_sell = ', len(sell))
    print('\tKol_good = ', kol_good)
    print('\tAll_money = ', all_money)
```

## Вариант с Моделькой

```python
data = pd.read_csv('/content/drive/My Drive/Data.csv', index_col = 0, usecols=['time', 'means']).reset_index()
data.time = pd.to_datetime(data.time)
data.shape
```

```python
def create_dataset(dataset, look_back = 1, back_y = 0, del_col = -1, regression = False, window = 0, binary=True):
    if 'numpy' not in str(type(data)):
        dataset = np.array(dataset)
    if del_col != -1:
        dataset = np.delete(dataset, del_col, axis = 1)
    scaler = MinMaxScaler(feature_range=(-1, 1))
    if regression:
        window = 0
        dataset = scaler.fit_transform(dataset)
    else:
        back_y = 0
        # dataset = scaler.fit_transform(dataset)
    # dataset[:, :-1] = StandardScaler().fit_transform(dataset[:, :-1])
    dataX, dataY = np.zeros((dataset.shape[0]-look_back-window, look_back, 1)), np.zeros((dataset.shape[0]-look_back-window, 1))
    k = 0
    for i in range(look_back, len(dataset) - back_y - window ):
        if regression:
            a = dataset[i-look_back:i]
            dataX[k] = a
            dataY[k] = dataset[i:i + back_y+1]
        else:
            a = dataset[i-look_back:i]
            dataX[k] = a
            if binary:
                if dataset[i] > dataset[i - 1]:
                    dataY[k] = [1]
                else:
                    dataY[k] = [0]
            else:
                if dataset[i - window - 1:i + window].max() == dataset[i - 1]:
                    dataY[k-2:k+2+1] = [1]
                elif dataset[i - window - 1:i+window].min() == dataset[i - 1]:
                    dataY[k-2:k+2+1] = [2]
        k+=1
    dataX = np.array(dataX, dtype=float)
    dataY = np.array(dataY, dtype = float if regression else int)
    if not regression:
        if binary:
            dataY = np.eye(2)[dataY[:, 0]]
        else:    
            dataY = np.eye(3)[dataY[:, 0]]
    return dataX, dataY
```

```python
%%time
window = 20
dataset = np.array(data.rolling(window = 20).mean())[window-1::10]
X, y = create_dataset(data, look_back=20, del_col=0, window = window, regression=False, binary = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, shuffle=True)
X_train.shape, y_train.shape
```

```python
def build_model():
    input = tf.keras.layers.Input((20, 1))
    out = LSTM(50)(input)
    out = Dense(2, activation='softmax')(out)
    return tf.keras.Model(inputs=input, outputs=out)

model = build_model()
```

```python
optim = tf.keras.optimizers.Adam()
loss = tf.keras.losses.CategoricalCrossentropy()
try:
    auc
except NameError:
    auc = tf.keras.metrics.AUC()
model.compile(optimizer=optim, loss = loss, metrics = [auc, 'accuracy'])
```

```python
class_weight = {
    0: len(y)/y.sum(axis=0)[0],
    1: len(y)/y.sum(axis=0)[1],
    # 2: len(y)/y.sum(axis=0)[2],
}
class_weight
```

```python
NUM_EPOCHS = 20
BATCH_SIZE = 10
```

```python
hist = model.fit(X_train, y_train, batch_size=BATCH_SIZE
                 , epochs=NUM_EPOCHS, validation_data = (X_test, y_test)
                 , class_weight = class_weight
                 )
```

```python
for i in range(NUM_EPOCHS):
    print('Epoch = ', i, end = ' ')
    hist = np.zeros((len(X_train), 3))
    for j in range(0, len(X_train)//BATCH_SIZE):
        h = model.train_on_batch(x=X_train[j*BATCH_SIZE:(j+1)*BATCH_SIZE]
                                , y = y_train[j*BATCH_SIZE:(j+1)*BATCH_SIZE], class_weight=class_weight)
        hist[j] = h
    print(np.array(hist).mean(axis=0))
    model.reset_states()
```

```python
hist = []
for j in range(0, len(X_test), BATCH_SIZE):
    h = model.test_on_batch(x=X_test[i:i+BATCH_SIZE], y = y_test[i:i+BATCH_SIZE])
    hist.append(h)
print(np.array(hist).mean(axis=0))
```
